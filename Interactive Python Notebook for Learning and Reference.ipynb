```python
import tkinter as tk
import fitz  

def display_pdf_content(pdf_file):
    doc = fitz.open(pdf_file)
    content = []

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        content.append(page.get_text())

    return "\n".join(content)

def update_display(selected_option):
    pdf_mapping = {
        "Python": "python.pdf",                                      # give your path where the pdf is present
        "SQL": "SQL.pdf",
        "Data Science": "data_science.pdf",
        "Power BI": "Power_BI.pdf",
        "Statistics": "statistics.pdf",
        "List": "list.pdf",
        "Tuple": "tuple.pdf",
        "Set": "set.pdf",
        "Dictionary": "dict.pdf",
        "If-Else": "if_else.pdf",
        "Function": "function.pdf",
        "Class": "class.pdf",
        "Oops": "oops.pdf"
    }

    pdf_file = pdf_mapping.get(selected_option)

    if pdf_file:
        content = display_pdf_content(pdf_file)
        display_text.delete(1.0, tk.END)  # Clear the text widget
        display_text.insert(tk.END, content)

    if selected_option == "Python":
        hide_other_buttons()
        show_python_suboptions()
    else:
        show_all_buttons()

def hide_other_buttons():
    for widget in button_frame.winfo_children():
        if widget["text"] != "Python":
            widget.pack_forget()

def show_all_buttons():
    for widget in button_frame.winfo_children():
        widget.pack(side="top", padx=10, pady=5)  # Arrange buttons vertically

def show_python_suboptions():
    suboptions = ["List", "Tuple", "Set", "Dictionary", "If-Else", "Function", "Class", "Oops"]
    
    # Create a new frame for the suboptions buttons
    suboptions_frame = tk.Frame(button_frame, bg="#272727")
    suboptions_frame.pack(side="left", padx=10, anchor="w")

    for option in suboptions:
        button = tk.Button(suboptions_frame, text=option, font=("Arial", 12), command=lambda o=option: update_display(o), fg="white", bg="#007BFF")
        button.pack(fill="x")

root = tk.Tk()
root.title("Tanmaya")
root.configure(bg="#272727") 

label = tk.Label(root, text="Welcome to Notebook", font=("Arial", 40), fg="white", bg="#272727")
label.pack(pady=10)

options = ["Python", "SQL", "Data Science", "Power BI", "Statistics"]

button_frame = tk.Frame(root, bg="#272727")  
button_frame.pack(side="left", padx=10, pady=10, anchor="w")

for option in options:
    button = tk.Button(button_frame, text=option, font=("Arial", 12), command=lambda o=option: update_display(o), fg="white", bg="#007BFF")
    button.pack(fill="x")

display_text = tk.Text(root, height=35, width=60, bg="#1E1E1E", fg="white", font=("Arial", 16))  
display_text.pack(side="left", padx=10, pady=10)

root.mainloop()
```
